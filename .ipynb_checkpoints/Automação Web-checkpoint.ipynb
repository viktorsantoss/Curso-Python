```python
#importando as bibliotecas
#biblioteca do selenium (internet)
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd #base de dados
from time import sleep 

#bibliotecas do email
import smtplib #protocolo de envio de e-mail para o servidor
#o pacote MIME é o padrão do e-mail
from email.mime.multipart import MIMEMultipart #instância que cria o e-mail
from email.mime.text import MIMEText #adicionar o texto no escopo do e-mail
from email.mime.base import MIMEBase #adicionar o arquivo como anexo
from email.encoders import encode_base64 #codificar o arquivo em base 64

from pathlib import Path

options = webdriver.ChromeOptions()
options.add_argument('start-maximized')
navegador = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
#importando a base de dados
base_dados = pd.read_excel('buscas.xlsx')
base_dados
```

### Funções do google e buscape

```python
def google(nome_produto, valor_minimo, valor_maximo, termos_banidos, termos_produtos): 
    resposta = []
    navegador.get('https://www.google.com.br/')
    #pesquisando pelo produto
    navegador.find_element(By.XPATH, '//*[@id="APjFqb"]').send_keys(nome_produto, Keys.ENTER)
    #indo para o shopping
    elementos_pesquisa = navegador.find_elements(By.CLASS_NAME, 'hdtb-mitem')
    for pesquisa in elementos_pesquisa:
        if 'Shopping' in pesquisa.text:
            pesquisa.click()
            break

    #filtrar pelo valor máximo e mínimo
    sleep(5)
    navegador.find_element(By.NAME, 'lower').send_keys(valor_minimo) #envia o valor mínimo
    navegador.find_element(By.NAME, 'upper').send_keys(valor_maximo) #envia o valor máximo
    navegador.find_element(By.CLASS_NAME, 'sh-dr__prs').click()

    #lista contendo todos os produtos
    lista_produtos = navegador.find_elements(By.CLASS_NAME, 'i0X6df')
    for produto in lista_produtos:    
        #nome do produto
        nome = produto.find_element(By.CLASS_NAME, 'EI11Pd').text
        
        #verificar se tem os termos banidos no nome
        situacao_termos_banidos = False
        for palavra in termos_banidos:
            if palavra.lower() in nome.lower():
                situacao_termos_banidos = True
                
        #verificar se as palavras dos nomes_produtos está no nome
        situacao_nome_produtos = True
        for termo in termos_produtos:
            if not termo.lower() in nome.lower():
                situacao_nome_produtos = False
        
        #no caso do iphone
        if 'iphone' in nome_produto.lower():
            indice = nome_produto.find('2')
            nome_produto_verificação = nome_produto[:indice + 1] #pega o nome iphone 12
            #verifica agora se o "iphone 12" está contido no nome
            if not nome_produto_verificação.lower() in nome.lower():
                situacao_nome_produtos = False

        if not situacao_termos_banidos and situacao_nome_produtos:
            #valor do produto
            valor = produto.find_element(By.CLASS_NAME, 'a8Pemb').text
            # link do produto
            elemento_filho = produto.find_element(By.CLASS_NAME, 'XrAfOe') #pega a classe que está contida no link
            link = elemento_filho.find_element(By.XPATH, '..').get_attribute('href') #pega o pai da classe, que é o link
            #nome da loja
#             loja = produto.find_element(By.CLASS_NAME, 'aULzUe').text     
            if not "impostos" in valor: #tira produtos que contém o valor + impostos
                resposta.append((nome, valor, link))
    return resposta



def buscape(nome_produto, valor_minimo, valor_maximo, termos_banidos, termos_produtos):
    resposta = []
    navegador.get('https://www.buscape.com.br/')
    #pesquisando o produto
    navegador.find_element(By.TAG_NAME, 'input').send_keys(nome_produto, Keys.ENTER)
    #pesquisando a faixa de valores
    sleep(5)
    campo_inputs = navegador.find_elements(By.CLASS_NAME, 'RefinementPrice_Input__DaDQH')
    campo_inputs[0].find_element(By.TAG_NAME, 'input').send_keys(Keys.CONTROL, 'a', Keys.DELETE)
    campo_inputs[0].find_element(By.TAG_NAME, 'input').send_keys(f'{valor_minimo:.2f}')

    campo_inputs[1].find_element(By.TAG_NAME, 'input').send_keys(Keys.CONTROL, 'a', Keys.DELETE)
    campo_inputs[1].find_element(By.TAG_NAME, 'input').send_keys(f'{valor_maximo:.2f}')
    sleep(3)
    navegador.find_element(By.CLASS_NAME, 'RefinementPrice_Button__azwX5').click()
    sleep(5)
    produtos = navegador.find_elements(By.CLASS_NAME, 'SearchCard_ProductCard__1D3ve')
    for produto in produtos:
        try:
            nome = produto.find_element(By.TAG_NAME, 'h2').text
            
            #verificar se tem os termos banidos no nome
            situacao_termos_banidos = False
            for palavra in termos_banidos:
                if palavra.lower() in nome.lower():
                    situacao_termos_banidos = True
                    
            #verificar se as palavras dos nomes_produtos está no nome             
            situacao_nome_produtos = True
            for termo in termos_produtos:
                if not termo.lower() in nome.lower():
                    situacao_nome_produtos = False
                    
            #no caso do iphone
            if 'iphone' in nome_produto.lower():
                indice = nome_produto.find('2')
                nome_produto_verificação = nome_produto[:indice + 1] #pega o nome iphone 12
                #verifica agora se o "iphone 12" está contido no nome
                if not nome_produto_verificação.lower() in nome.lower():
                    situacao_nome_produtos = False
                    
                    
            if not situacao_termos_banidos and situacao_nome_produtos:
                valor = produto.find_element(By.CLASS_NAME, 'Text_MobileHeadingS__XS_Au').text
                link = produto.find_element(By.TAG_NAME, 'a').get_attribute('href')
                resposta.append((nome, valor, link))
        except:
            continue
    return resposta
            
```

### Percorrendo os produtos da base de dados

```python
for indice in base_dados.index:    
    nome_produto = base_dados.loc[indice, 'Nome']
    valor_minimo = float(base_dados.loc[indice, 'Preço mínimo'])
    valor_maximo = float(base_dados.loc[indice, 'Preço máximo'])
    termos_banidos = base_dados.loc[indice, 'Termos banidos'].split(' ')
    termos_produtos =  nome_produto.split(' ')
    tabela_final = pd.DataFrame()
    valor_google = google(nome_produto, valor_minimo, valor_maximo, termos_banidos, termos_produtos)
    if len(valor_google) > 0:
        df_google = pd.DataFrame(valor_google, columns=['Produto', 'preço', 'Link'])
#         display(df_google)
    else:
        df_google = None

    valor_buscape = buscape(nome_produto, valor_minimo, valor_maximo, termos_banidos, termos_produtos)
    if len(valor_buscape) > 0:
        df_buscape = pd.DataFrame(valor_buscape, columns=['Produto', 'preço', 'Link'])
#         display(df_buscape)
    else:
        df_buscape = None
    if len(df_google.index) > 0 or len(df_buscape.index) > 0:
        tabela_final  = pd.concat([df_google, df_buscape], ignore_index=True) 
        tabela_final.sort_values('preço', ascending=True, inplace=True) #organiza os valores do menor para o maior
        tabela_final.to_excel(f'{nome_produto}.xlsx', index=False)

navegador.quit() #saindo do navegador    
```

### Enviando para o e-mail

```python
#inicializando as variáveis
host = 'smtp.gmail.com' #servidor do g-mail
port = '587'
usuario = '' #email que vai enviar a mensagem
senha = '' #senha gerada na parte de segurança

#criando o e-mail
mensagem = MIMEMultipart()
mensagem['From'] = usuario
mensagem['To'] = usuario
mensagem['Subject'] = 'Produto(s) Encontrado(s) na faixa de preço desejada'
texto = f"""
 <p>Prezados,</p>
    <p>Encontrou-se alguns produtos em oferta dentro da faixa de preço desejada. Segue tabela com detalhes</p>
    <p>Qualquer dúvida estou à disposição</p>
    <p>Att., Victor 
    </p>
    """

mensagem.attach(MIMEText(texto, 'html'))
caminho = Path.cwd() 
for anexo in caminho.iterdir():
    nome = anexo.name
    if 'xlsx' in nome and not 'buscas' in nome:
        with open(anexo, 'rb') as arquivo:
            anexo = MIMEBase('application', 'octet-stream')
            anexo.set_payload(arquivo.read())
            encode_base64(anexo)
            anexo.add_header('content-disposition', 'attachment', filename=f'{nome}')
            mensagem.attach(anexo)

with smtplib.SMTP(host, port) as servidor:
    servidor.starttls()
    servidor.login(usuario, senha)
    servidor.sendmail(mensagem['From'], mensagem['To'], mensagem.as_string().encode('utf-8'))
```
